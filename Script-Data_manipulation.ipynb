```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

```python
datasets = pd.read_csv(pwd + "\Data-Washington_Fish_Harvest_Data_1974-2012.csv")
```

```python
datasets
```

```python
#Looking at the total amount of harvest over the years
totalFishByYear = datasets.copy()
totalFishByYear = totalFishByYear[["Year", "Total by Year"]]
totalFishByYear =totalFishByYear.dropna()

x = totalFishByYear.Year
y = totalFishByYear["Total by Year"] / 1000000
plt.plot(x,y)
plt.title("Total Fish Harvested by Year")
plt.xlabel("Year")
plt.ylabel("Fish harvested (millions)")

z= np.polyfit(x,y,1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show
```

```python
# Next I want to explore the types of fisher distributions by year
from matplotlib.pyplot import subplots


fisherTypeTotal = datasets.copy()
fisherTypeTotal = fisherTypeTotal[["Year", "Fisher Type", "Total by Fisher Type"]] 
fisherlabel = ["Sport", "Non-Treaty", "Treaty"]
year1 = fisherTypeTotal.iloc[[0,1,2]].groupby(["Year"]).plot(kind= 'pie', title= "Harvest type of 1974" ,y="Total by Fisher Type",ylabel="", labels = fisherlabel,autopct='%1.1f%%', explode=[0,0,0.05])
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
year2 = fisherTypeTotal.iloc[[3,4,5]].groupby(["Year"]).plot.pie(title= "Harvest type of 1975" ,y="Total by Fisher Type",ylabel="", labels = fisherlabel,autopct='%1.1f%%', explode=[0,0,0.05])
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
year3 = fisherTypeTotal.iloc[[6,7,8]].groupby(["Year"]).plot.pie(title= "Harvest type of 1975" ,y="Total by Fisher Type",ylabel="", labels = fisherlabel,autopct='%1.1f%%', explode=[0,0,0.05])
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')

year4 = fisherTypeTotal.iloc[[105,106,107]].groupby(["Year"]).plot.pie(title= "Harvest type of 2009" ,y="Total by Fisher Type",ylabel="", labels = fisherlabel,autopct='%1.1f%%', explode=[0,0,0.05])
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
year5 = fisherTypeTotal.iloc[[108,109,110]].groupby(["Year"]).plot.pie(title= "Harvest type of 2010" ,y="Total by Fisher Type",ylabel="", labels = fisherlabel,autopct='%1.1f%%', explode=[0,0,0.05])
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
year6 = fisherTypeTotal.iloc[[111,112,113]].groupby(["Year"]).plot.pie(title= "Harvest type of 2011" ,y="Total by Fisher Type",ylabel="", labels = fisherlabel,autopct='%1.1f%%', explode=[0,0,0.05])
plt.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
```
